# Central Limit Theorem Illustrations

```python
import numpy as np
import pandas as pd
import math
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
sns.set(style='ticks', palette='Set2')
%matplotlib inline
```

```python
samples = []
samples_all = []

number_in_sum = [2,3,5,7,10,1000]
for size in number_in_sum:
    for number_in_sample in range(1,10000):
        tmp = np.random.random_integers(0, 1, size=size)
        tmp[(tmp == 1)] = 1
        tmp[(tmp == 0)] = -1
        #tmp = np.random.randn(1)
        samples.append(sum(tmp)/math.sqrt(size))
    samples_all.append(samples)
    samples = []
df = pd.DataFrame({ 'n0' : samples_all[0], 'n1' : samples_all[1], 'n2' : samples_all[2],
                    'n3' : samples_all[3], 'n4' : samples_all[4], 'n5' : samples_all[5] })

figsize(12,8)
fig, axes = plt.subplots(nrows=2, ncols=3)
df.n0.hist(color='blue',ax=axes[0][0], alpha=0.9, bins=1000)
df.n1.hist(ax=axes[0][1],color='r',bins=1000)
df.n2.hist(ax=axes[0][2],color='r',bins=1000)
df.n3.hist(ax=axes[1][0],color='r',bins=1000)
df.n4.hist(ax=axes[1][1],color='r',bins=1000)
df.n5.hist(ax=axes[1][2],color='r',bins=1000)

axes[0][0].set_title('N = '+str(number_in_sum[0]))
axes[0][1].set_title('N = '+str(number_in_sum[1]))
axes[0][2].set_title('N = '+str(number_in_sum[2]))
axes[1][0].set_title('N = '+str(number_in_sum[3]))
axes[1][1].set_title('N = '+str(number_in_sum[4]))
axes[1][2].set_title('N = '+str(number_in_sum[5]))
plt.suptitle(r'$Z = \frac{1}{\sqrt{N}} \sum_{i=1}^N X_i$')
plt.show()
```

# What does this mean and how is it useful?

```python
# let's look at expectations and coin flips
```

```python
def number_of_heads(n,p): # returns the number of heads out of n, of a biased coin
    h = np.random.rand(n) 
    H = 1.0*(h <= p)
    return sum(H)
```

```python
# How many heads do we expect? Let's try this:
n = 1000
H = number_of_heads(n,0.5)
print H
```

```python
# How likely is it to see 470 heads? How about 600 heads?
# And how many experiments would we need to see just 100 heads?
```

```python
# Let's repeat this experiment and plot the results
k = 500
heads = np.zeros(k)
for i in np.arange(k):
    heads[i] = number_of_heads(n,0.5)
p = plt.hist(heads,50)
plt.title("Number of Heads")
plt.show()
```

```python
# variance of the H/T process: Var(X) = E[(X - EX)^2] = E[X^2]-E[X]^2 = p - p^2 = p(1-p)
# so standard dev = sqrt(p(1-p))
# let's check numerically:
#
# (X - EX)^2
def bern_dev(p):
    b = 1.0*(np.random.rand(1) < p)
    return (b - p)**2
p = 0.35
var = p*(1-p)
print var
average = 0
for i in np.arange(100000):
    average = average + bern_dev(p)/100000
print average
```

```python
# Let's use the CLT to figure out the answers above.
# CLT says: sqrt(n)(Avg Number of Heads - p)/stddev ~ N(0,1)
#
# Let's check the CLT:
#
k = 5000
n = 10000
p = 0.5
std_dev = math.sqrt(p*(1-p))
heads = np.zeros(k)
for i in np.arange(k):
    heads[i] = number_of_heads(n,p)
# normalize as above:
normalized_heads = math.sqrt(n)*(heads/n - p)/std_dev
```

```python
mean = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(-3,3,100)
plt.plot(x,mlab.normpdf(x,mean,sigma))
sns.distplot(normalized_heads)
```
